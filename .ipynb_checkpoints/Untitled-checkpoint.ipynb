```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
 
%matplotlib inline
```

```python
df = pd.read_csv('fcc-forum-pageviews.csv',
                parse_dates=['date'],
                index_col='date')
df.head()
```

```python
df.info()
```

#### q2
Clean the data by filtering out days when the page views were in the top 2.5% of the dataset or bottom 2.5% of the dataset.

```python
cond = ( df['value'] <= df['value'].quantile(0.975) ) & \
       ( df['value'] >= df['value'].quantile(0.025) )
df = df.loc[cond]
```

#### q3
Create a `draw_line_plot` function that uses Matplotlib to draw a line chart similar to "examples/Figure_1.png". The title should be `Daily freeCodeCamp Forum Page Views 5/2016-12/2019`. The label on the x axis should be `Date` and the label on the y axis should be `Page Views`.

```python
fig, axes = plt.subplots(figsize=(10, 4))
axes.plot(df.index, df['value'], color='red')
axes.set_xlabel('Date')
axes.set_ylabel('Page Views')
axes.set_title('Daily freeCodeCamp Forum Page Views 5/2016-12/2019')
```

#### q4
Create a `draw_bar_plot` function that draws a bar chart similar to "examples/Figure_2.png". It should show average daily page views for each month grouped by year. The legend should show month labels and have a title of `Months`. On the chart, the label on the x axis should be `Years` and the label on the y axis should be `Average Page Views`.

```python
df_bar_plot = df.copy()

df_bar_plot['month_num'] = df_bar_plot.index.month
df_bar_plot['year'] = df_bar_plot.index.year

df_bar_plot.head()
```

```python
df_bar = df_bar_plot.groupby(by=['year', 'month_num']).mean().sort_values(by=['year','month_num'])
df_bar.head(10)
```

```python
labels = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 
          'September', 'October', 'November', 'December']

fig, ax = plt.subplots(figsize=(12, 6))
df_bar.unstack(level='month_num').plot(kind='bar', ax=ax, legend=True)
ax.legend(title='Months', labels=labels)
ax.set_xlabel('Years')
ax.set_ylabel('Average Page Views')
```

#### q5
Create a `draw_box_plot` function that uses Seaborn to draw two adjacent box plots similar to "examples/Figure_3.png".

These box plots should show how the values are distributed within a given year or month and how it compares over time. The title of the first chart should be `Year-wise Box Plot (Trend)` and the title of the second chart should be `Month-wise Box Plot (Seasonality)`. 

Make sure the month labels on bottom start at Jan and the x and y axis are labeled correctly. The boilerplate includes commands to prepare the data.

```python
df_box = df.copy()
df_box.reset_index(inplace=True)
df_box['year'] = [d.year for d in df_box.date]
df_box['month'] = [d.strftime('%b') for d in df_box.date]
```

```python
df_box
```

```python
months_order  = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
plot_objects = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

fig, ((ax1, ax2)) = plot_objects
sns.boxplot(data=df_box, x='year', y='value', ax=ax1, orient='v')
sns.boxplot(data=df_box, x='month', y='value', ax=ax2, order=months_order, orient='v')
ax1.set_title('Year-wise Box Plot (Trend)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Page Views')
ax2.set_title('Month-wise Box Plot (Seasonality)')
ax2.set_xlabel('Month')
ax2.set_ylabel('Page Views')
```
